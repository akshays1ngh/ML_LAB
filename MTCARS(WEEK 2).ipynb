```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mtcars.csv')
print("DataFrame: ")
df
```

```python
#Q1)  Tabulate number of cylinders in the dataset.
cylinders=df['cyl'].value_counts()
print("Cylinders are: ",cylinders)
print("Total number of cylinders: ",sum(cylinders))
```

```python
#Q2) Find the five number summary of milespergallon.
#1st way
df['mpg'].describe()
```

```python
#2nd way
from numpy import percentile
quartiles=percentile(df['mpg'],[25,50,75])

print("5 number summary is: ")
print("Minimum is: ",df['mpg'].min())
print("1st quartile is: ",quartiles[0])
print("2nd quartile is: ",quartiles[1])
print("3rd quartile is: ",quartiles[2])
print("Maximum is: ",df['mpg'].max())
```

```python
# Q3) Draw a histogram for hp.

plt.hist(df['hp'],bins=10,edgecolor='yellow')
plt.xlabel("Horsepower")
plt.title("Histogram for HP")
plt.show()
```

```python
# Q4) Box plot the miles per gallon.

plt.boxplot(df['mpg'])
plt.ylabel('MPG')
plt.title('BOXPLOT FOR MILES PER GALLON')
plt.show()
```

```python
#Q5)  Find the avg weight of all cars.

print("Average weight of cars is: ",df['wt'].mean())
```

```python
#Q6) Find the car with the minimum displacement.

print("Car with min displacement is: ")
print(df.loc[df['disp'].idxmin(),'model'])
```

```python
#Q7) . Find the car with the maximum qsec.

print("Car with max qsec is: ")
print(df.loc[df['qsec'].idxmax(),'model'])
```

```python
# Q8)  Find the median horse power and find the car with the highest fuel .

print("Median horsepower is: ")
print(df['hp'].median())
print("Car with highest fuel is: ")
print(df.loc[df['mpg'].idxmax(),'model'])
```

```python
# Q9) Find the car with the lowest fuel efficiency.

print("Car with lowest fuel is: ")
print(df.loc[df['mpg'].idxmin(),'model'])
```

```python
#Q10)  Find the car with best hp.
print("Car with best hp is: ")
print(df.loc[df['hp'].idxmax(),'model'])
```

```python
#Q11)  Tabulate mpg for different number of gears.

gears=df.groupby(by='gear')

for gear,group in gears:
    print("MPG for gear type: ",gear)
    print(group['mpg'].to_list())
    print()

avg_mean=gears['mpg'].mean()

print("Average mean of respective gears is: ")
print(avg_mean)
```

```python
#Q11)  Do side by side boxplot of mpg for cars with am (automatic transmission) Vs cars with Manual transmission.

#for automatic transmission, am = 0 and for manual transmission, am = 1

#creating a list of dataframes for cars with am = 0 and am = 1
data = [df[df['am'] == 0]['mpg'], df[df['am'] == 1]['mpg']]

#create a boxplot for the data
fig = plt.figure(figsize=(10, 5))
bp = plt.boxplot(data, patch_artist=True, notch=False)

#add colors to the boxplot
colors = ['pink', 'lightblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

#add labels and title
plt.xticks([1, 2], ['Manual', 'Automatic'])
plt.ylabel('Miles per gallon')
plt.title('MPG for cars with manual VS automatic transmission')

#show the plot
plt.show()
```
